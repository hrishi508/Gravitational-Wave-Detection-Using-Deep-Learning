# Noise

```python
for i in tqdm(range(100)):
    # The color of the noise matches a PSD which you provide
    flow = 30.0
    delta_f = 1.0 / 16
    flen = int(2048 / delta_f) + 1
    psd = pycbc.psd.aLIGOZeroDetHighPower(flen, delta_f, flow)

    # Generate 4 seconds of noise at 4096 Hz
    delta_t = 1.0 / 4096
    tsamples = int(4 / delta_t)
    noise = pycbc.noise.noise_from_psd(tsamples, delta_t, psd)
    noise = TimeSeries.from_pycbc(noise)
    plt.plot(noise)

plt.title('ALIGOZeroDetHighPower Noise')
plt.savefig(noise_directory + 'ALIGOZeroDetHighPower.png')
```

```python
for i in tqdm(range(100)):
    noise = TimeSeries(np.random.normal(scale=.1, size=16384), sample_rate=4096)
    plt.plot(noise)
plt.title('Gaussian Noise')
plt.savefig(noise_directory + 'Gaussian.png')
```
